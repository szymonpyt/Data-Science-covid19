```python
import pandas as pd

df = pd.read_csv('covid19_confirmed_global.csv')

df
```

```python
df = df.drop(['Province/State', 'Lat', 'Long'], axis=1)
df
```

```python
df['Country/Region'].unique()
```

```python
countries_to_plot = ['Poland', 'US', 'Italy']

import matplotlib.pyplot as plt

# US Date format: 'MM/DD/YYYY'
x_values = list(df)[1:]

for country in countries_to_plot:
    country_df = df[df['Country/Region'] == country ]
    y_values = [int(country_df[date]) for date in x_values]
    plt.plot(x_values, y_values, label=country)

plt.style.use('fivethirtyeight')
plt.title('Confirmed Cases Over Time')
skip = len(x_values)//5 # Helps ensure we don't add too many date tick marks
plt.xticks(x_values[::skip])
plt.xlabel('Dates')
plt.ylabel('Confirmed cases')
plt.legend()
plt.show()
```
